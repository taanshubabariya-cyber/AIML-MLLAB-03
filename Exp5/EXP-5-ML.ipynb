```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
```

```python
# Load real dataset
data = load_breast_cancer()

X = data.data
y = data.target

print("Classes:", data.target_names)
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Logistic Regression
lr = LogisticRegression(max_iter=5000)
lr.fit(X_train, y_train)

pred_lr = lr.predict(X_test)
print("Logistic Regression TestAccuracy:", accuracy_score(y_test, pred_lr))

pred_lr_train = lr.predict(X_train)
print("Logistic TrainAccuracy:", accuracy_score(y_train, pred_lr_train))
```

```python
# Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

pred_dt = dt.predict(X_test)
print("Decision Tree TestAccuracy:", accuracy_score(y_test, pred_dt))

pred_dt_train = dt.predict(X_train)
print("Decision Tree TrainAccuracy:", accuracy_score(y_train, pred_dt_train))
```

```python
# KNN
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

pred_knn = knn.predict(X_test)
print("KNN TestAccuracy:", accuracy_score(y_test, pred_knn))

pred_knn_train = knn.predict(X_train)
print("KNN TrainAccuracy:", accuracy_score(y_train, pred_knn_train))
```

```python
# Naive Bayes
nb = GaussianNB()
nb.fit(X_train, y_train)

pred_nb = nb.predict(X_test)
print("Naive Bayes TestAccuracy:", accuracy_score(y_test, pred_nb))

pred_nb_train = nb.predict(X_train)
print("Naive Bayes TrainAccuracy:", accuracy_score(y_train, pred_nb_train))
```

```python
# MLP Classifier
mlp = MLPClassifier(max_iter=5000)
mlp.fit(X_train, y_train)

pred_mlp = mlp.predict(X_test)
print("MLP TestAccuracy:", accuracy_score(y_test, pred_mlp))

pred_mlp_train = mlp.predict(X_train)
print("MLP TrainAccuracy:", accuracy_score(y_train, pred_mlp_train))
```

```python
# Confusion Matrix + Type-I and Type-II Error (All 5 Models)

# 1) Logistic Regression
cm_lr = confusion_matrix(y_test, pred_lr)
print("\nLogistic Regression Confusion Matrix:\n", cm_lr)

TN1, FP1, FN1, TP1 = cm_lr.ravel()

type1_lr = FP1 / (FP1 + TN1)
type2_lr = FN1 / (FN1 + TP1)

print("Logistic Regression Type-I Error Rate:", type1_lr)
print("Logistic Regression Type-II Error Rate:", type2_lr)


# 2) Decision Tree
cm_dt = confusion_matrix(y_test, pred_dt)
print("\nDecision Tree Confusion Matrix:\n", cm_dt)

TN2, FP2, FN2, TP2 = cm_dt.ravel()

type1_dt = FP2 / (FP2 + TN2)
type2_dt = FN2 / (FN2 + TP2)

print("Decision Tree Type-I Error Rate:", type1_dt)
print("Decision Tree Type-II Error Rate:", type2_dt)


# 3) KNN
cm_knn = confusion_matrix(y_test, pred_knn)
print("\nKNN Confusion Matrix:\n", cm_knn)

TN3, FP3, FN3, TP3 = cm_knn.ravel()

type1_knn = FP3 / (FP3 + TN3)
type2_knn = FN3 / (FN3 + TP3)

print("KNN Type-I Error Rate:", type1_knn)
print("KNN Type-II Error Rate:", type2_knn)


# 4) Naive Bayes
cm_nb = confusion_matrix(y_test, pred_nb)
print("\nNaive Bayes Confusion Matrix:\n", cm_nb)

TN4, FP4, FN4, TP4 = cm_nb.ravel()

type1_nb = FP4 / (FP4 + TN4)
type2_nb = FN4 / (FN4 + TP4)

print("Naive Bayes Type-I Error Rate:", type1_nb)
print("Naive Bayes Type-II Error Rate:", type2_nb)


# 5) Neural Network (MLP)
cm_mlp = confusion_matrix(y_test, pred_mlp)
print("\nNeural Network (MLP) Confusion Matrix:\n", cm_mlp)

TN5, FP5, FN5, TP5 = cm_mlp.ravel()

type1_mlp = FP5 / (FP5 + TN5)
type2_mlp = FN5 / (FN5 + TP5)

print("Neural Network (MLP) Type-I Error Rate:", type1_mlp)
print("Neural Network (MLP) Type-II Error Rate:", type2_mlp)
```

```python
precision = precision_score(y_test, pred_dt)
recall = recall_score(y_test, pred_dt)
f1 = f1_score(y_test, pred_dt)
roc = roc_auc_score(y_test, pred_dt)
print(f"Precision: {precision}")
print(f"Recall: {recall}")  
print(f"F1 Score: {f1}")
print(f"ROC AUC Score: {roc}")
```
